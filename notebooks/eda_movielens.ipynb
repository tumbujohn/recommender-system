# Collaborative Filtering: Step-by-Step

This section demonstrates both user-based and item-based collaborative filtering for movie recommendations using the MovieLens dataset. We will:
- Prepare the user-item matrix
- Compute similarities
- Generate recommendations for a sample user


```python
# Prepare the user-item matrix
user_item_matrix = ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)
user_item_matrix.head()
```

## User-Based Collaborative Filtering

We will compute user-user similarities and recommend movies liked by similar users.

```python
# %pip install scikit-learn #Required for similarity calculations
# Import necessary libraries for similarity calculations if not already installed in environment. 
# In my case, it is already installed in my environment. @tumbujohn \miniconda3\envs\recommender\

from sklearn.metrics.pairwise import cosine_similarity

# Compute user-user similarity matrix
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
user_similarity_df.head()
```

```python
# Recommend movies for a sample user using user-based CF
sample_user = 1  # Change this to try other users

# Find the most similar users (excluding the user themself)
similar_users = user_similarity_df[sample_user].drop(sample_user).sort_values(ascending=False)

# Get movies rated by similar users but not by the sample user
user_movies = set(ratings[ratings['userId'] == sample_user]['movieId'])
recommendations = []
for neighbor in similar_users.index:
    neighbor_movies = ratings[ratings['userId'] == neighbor][['movieId', 'rating']]
    for _, row in neighbor_movies.iterrows():
        if row['movieId'] not in user_movies:
            recommendations.append((row['movieId'], row['rating']))
    if len(recommendations) > 100:
        break

# Aggregate and sort recommendations by average rating
import collections
rec_counter = collections.defaultdict(list)
for movie_id, rating in recommendations:
    rec_counter[movie_id].append(rating)
rec_list = [(movie_id, np.mean(ratings)) for movie_id, ratings in rec_counter.items()]
rec_list = sorted(rec_list, key=lambda x: x[1], reverse=True)[:5]

# Add movie titles
rec_df = pd.DataFrame(rec_list, columns=['movieId', 'predicted_rating'])
rec_df = rec_df.merge(movies, on='movieId', how='left')
display(rec_df[['movieId', 'title', 'predicted_rating']])
```

## Item-Based Collaborative Filtering

We will compute item-item similarities and recommend movies similar to those the user has already liked.

```python
# Compute item-item similarity matrix
item_item_matrix = user_item_matrix.T  # Transpose for item-based
item_similarity = cosine_similarity(item_item_matrix)
item_similarity_df = pd.DataFrame(item_similarity, index=item_item_matrix.index, columns=item_item_matrix.index)
item_similarity_df.head()
```

```python
# Recommend movies for a sample user using item-based CF
sample_user = 1  # Change this to try other users

# Get movies the user has rated
user_ratings = ratings[ratings['userId'] == sample_user][['movieId', 'rating']]

# For each movie the user has rated, find similar movies
scores = {}
for _, row in user_ratings.iterrows():
    movie_id = row['movieId']
    rating = row['rating']
    similar_scores = item_similarity_df[movie_id] * rating
    for sim_movie_id, score in similar_scores.items():
        if sim_movie_id not in user_ratings['movieId'].values:
            scores[sim_movie_id] = scores.get(sim_movie_id, 0) + score

# Get top 5 recommendations
rec_list = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:5]
rec_df = pd.DataFrame(rec_list, columns=['movieId', 'score'])
rec_df = rec_df.merge(movies, on='movieId', how='left')
display(rec_df[['movieId', 'title', 'score']])
```
